```python
import pandas as pd
import os
```

```python
# Đường dẫn thư mục chứa các file Excel
base_path = 'Data/Data raw'
```

```python
# Đọc từng file Excel và gán vào các biến tương ứng
data1 = pd.read_excel(f'{base_path}/2020-Vietnam.xlsx')
data2 = pd.read_excel(f'{base_path}/2021-Vietnam.xlsx')
data3 = pd.read_excel(f'{base_path}/2022-Vietnam.xlsx')
data4 = pd.read_excel(f'{base_path}/2023-Vietnam.xlsx')
data5 = pd.read_excel(f'{base_path}/2024-Vietnam.xlsx')
```

```python
# Kiểm tra danh sách các cột
print(data1.columns)
```

```python
# Lọc các cột có chữ "CĐKT." ở đầu
columns_to_select = ["Mã"] + [col for col in data1.columns if col.startswith("CĐKT.")]

# Nếu cột "Năm" tồn tại, thêm vào danh sách
if "Năm\nHợp nhất\nQuý: Hàng năm\nNăm: 2020\n" in data1.columns:
    columns_to_select.insert(1, "Năm\nHợp nhất\nQuý: Hàng năm\nNăm: 2020\n")  # Thêm "Năm" vào vị trí thứ hai

# Nếu tìm thấy cột có tên trong biến year_column, thêm vào danh sách
if year_column in data1.columns:
    columns_to_select.insert(1, year_column)

# Tạo DataFrame mới chỉ chứa các cột cần thiết
data1_CĐKT = data1[columns_to_select]

# Loại bỏ các ký tự không mong muốn khỏi tên cột
data1_CĐKT.columns = data1_CĐKT.columns.str.replace("\nHợp nhất\nQuý: Hàng năm\nNăm: 2020\n", "", regex=False)
data1_CĐKT.columns = data1_CĐKT.columns.str.replace("Đơn vị: Triệu VND", "", regex=False)
data1_CĐKT.columns = data1_CĐKT.columns.str.replace("CĐKT.", "", regex=False)
# Loại bỏ cột "Quý" nếu tồn tại
if "Quý" in data1_CĐKT.columns:
    data1_CĐKT = data1_CĐKT.drop(columns=["Quý"])

# Đổi tên cột còn lại thành "Năm"
if year_column in data1.columns:
    data1_CĐKT = data1_CĐKT.rename(columns={year_column: "Năm"})

# Thay thế các giá trị NaN hoặc trống bằng 0
data1_CĐKT = data1_CĐKT.fillna(0)

# Chuyển giá trị trong cột "Năm" thành số nguyên
if "Năm" in data1_CĐKT.columns:
    data1_CĐKT["Năm"] = data1_CĐKT["Năm"].astype(str).str.extract(r'(\d+)').astype(int)

# Hiển thị DataFrame sau khi xử lý
data1_CĐKT.head()
```

```python
# Lọc các cột có chữ "CĐKT." ở đầu
columns_to_select = ["Mã"] + [col for col in data4.columns if col.startswith("CĐKT.")]

# Nếu cột "Năm" tồn tại, thêm vào danh sách
if "Năm\nHợp nhất\nQuý: Hàng năm\nNăm: 2023\n" in data4.columns:
    columns_to_select.insert(1, "Năm\nHợp nhất\nQuý: Hàng năm\nNăm: 2023\n")  # Thêm "Năm" vào vị trí thứ hai

# Nếu tìm thấy cột có tên trong biến year_column, thêm vào danh sách
if year_column in data4.columns:
    columns_to_select.insert(1, year_column)

# Tạo DataFrame mới chỉ chứa các cột cần thiết
data4_CĐKT = data4[columns_to_select]

# Loại bỏ các ký tự không mong muốn khỏi tên cột
data4_CĐKT.columns = data4_CĐKT.columns.str.replace("\nHợp nhất\nQuý: Hàng năm\nNăm: 2023\n", "", regex=False)
data4_CĐKT.columns = data4_CĐKT.columns.str.replace("Đơn vị: Tỷ VND", "", regex=False)
data4_CĐKT.columns = data4_CĐKT.columns.str.replace("CĐKT.", "", regex=False)

# Loại bỏ cột "Quý" nếu tồn tại
if "Quý" in data4_CĐKT.columns:
    data4_CĐKT = data4_CĐKT.drop(columns=["Quý"])

# Đổi tên cột còn lại thành "Năm"
if year_column in data4.columns:
    data4_CĐKT = data4_CĐKT.rename(columns={year_column: "Năm"})

# Thay thế các giá trị NaN hoặc trống bằng 0
data4_CĐKT = data4_CĐKT.fillna(0)

# Chuyển giá trị trong cột "Năm" thành số nguyên
if "Năm" in data4_CĐKT.columns:
    data4_CĐKT["Năm"] = data4_CĐKT["Năm"].astype(str).str.extract(r'(\d+)').astype(int)
    
# Lấy danh sách các cột cần chuyển đổi (ngoại trừ "Mã" và "Năm")
cols_to_convert = [col for col in data4_CĐKT.columns if col not in ["Mã", "Năm"]]

# Nhân giá trị của các cột đó với 1000
data4_CĐKT[cols_to_convert] = data4_CĐKT[cols_to_convert].apply(lambda x: x * 1000)

# Hiển thị DataFrame sau khi xử lý
data4_CĐKT.head()
```
