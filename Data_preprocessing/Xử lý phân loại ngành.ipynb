```python
import pandas as pd
import os
```

```python
# Đường dẫn đến file Excel
file_path = 'Data/Data raw/Phan_loai_nganh.xlsx'
```

```python
# Đọc file Excel sử dụng pandas
data = pd.read_excel(file_path, header=8, nrows=1663)
```

```python
# Hiển thị dữ liệu đầu tiên của DataFrame
data.head(10)
```

```python
# Hiển thị dữ liệu đầu tiên của DataFrame
data.tail(10)
```

```python
# Đường dẫn đến thư mục lưu file cleaned
output_dir = 'Data/Data cleaned'
os.makedirs(output_dir, exist_ok=True)  # Tạo thư mục nếu chưa tồn tại
```

```python
# Đường dẫn file xuất ra
output_file_path = os.path.join(output_dir, 'Phan_loai_nganh(cleaned).xlsx')
```

```python
# Xuất dữ liệu ra file Excel
data.to_excel(output_file_path, index=False)
```

```python
print(f"File đã được xuất ra tại: {output_file_path}")
```
